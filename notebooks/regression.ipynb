```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
import graphviz

%matplotlib inline
```

```python
data = pd.read_csv('./data/processed_data.csv')
data_without_old = pd.read_csv('./data/data_2000_2020.csv')
```

```python
features = ['valence', 'acousticness', 'danceability',
            'duration_ms', 'energy', 'explicit', 'year', 'instrumentalness', 'key',
            'liveness', 'loudness', 'mode', 'speechiness', 'tempo']
features_2 = ['valence', 'acousticness', 'danceability',
            'duration_ms', 'energy', 'explicit', 'instrumentalness', 'key',
            'liveness', 'loudness', 'mode', 'speechiness', 'tempo']

X_1 = data[features]
y_1 = data.popularity

X_2 = data_without_old[features]
y_2 = data_without_old.popularity

X_3 = data_without_old[features_2]
y_3 = data_without_old.popularity
```

```python
train_X_1, test_X_1, train_y_1, test_y_1 = train_test_split(X_1, y_1, test_size=0.1, random_state=0)
train_X_2, test_X_2, train_y_2, test_y_2 = train_test_split(X_2, y_2, test_size=0.1, random_state=0)
train_X_3, test_X_3, train_y_3, test_y_3 = train_test_split(X_3, y_3, test_size=0.1, random_state=0)
```

```python
dtr_model_1 = DecisionTreeRegressor()
dtr_model_1.fit(train_X_1, train_y_1)

dtr_model_2 = DecisionTreeRegressor()
dtr_model_2.fit(train_X_2, train_y_2)

dtr_model_3 = DecisionTreeRegressor()
dtr_model_3.fit(train_X_3, train_y_3)
```

```python
prediction_1 = dtr_model_1.predict(test_X_1)
mae_1 = mean_absolute_error(test_y_1, prediction_1)
print('Mean absolute error of decision tree regression model #1: {}'.format(mae_1))

prediction_2 = dtr_model_2.predict(test_X_2)
mae_2 = mean_absolute_error(test_y_2, prediction_2)
print('Mean absolute error of decision tree regression model #2: {}'.format(mae_2))

prediction_3 = dtr_model_3.predict(test_X_3)
mae_3 = mean_absolute_error(test_y_3, prediction_3)
print('Mean absolute error of decision tree regression model #3: {}'.format(mae_3))
```

```python
dtr_1 = tree.export_graphviz(dtr_model_1, out_file=None, 
                                feature_names=features,  
                                class_names=y_1,
                                filled=True,
                                max_depth=5)
dtr_2 = tree.export_graphviz(dtr_model_2, out_file=None, 
                                feature_names=features,  
                                class_names=y_2,
                                filled=True,
                                max_depth=5)
dtr_3 = tree.export_graphviz(dtr_model_1, out_file=None, 
                                feature_names=features,  
                                class_names=y_3,
                                filled=True,
                                max_depth=5)

graph_1 = graphviz.Source(dtr_1, format="svg") 
graph_2 = graphviz.Source(dtr_2, format="svg") 
graph_3 = graphviz.Source(dtr_3, format="svg")
```

```python
rfr_model_1 = RandomForestRegressor()
rfr_model_1.fit(train_X_1, train_y_1)

rfr_model_2 = RandomForestRegressor()
rfr_model_2.fit(train_X_2, train_y_2)

rfr_model_3 = RandomForestRegressor()
rfr_model_3.fit(train_X_3, train_y_3)
```

```python
prediction_1 = rfr_model_1.predict(test_X_1)
mae_1 = mean_absolute_error(test_y_1, prediction_1)
print('Mean absolute error of random forest regression model #1: {}'.format(mae_1))

prediction_2 = rfr_model_2.predict(test_X_2)
mae_2 = mean_absolute_error(test_y_2, prediction_2)
print('Mean absolute error of random forest regression model #2: {}'.format(mae_2))

prediction_3 = rfr_model_3.predict(test_X_3)
mae_3 = mean_absolute_error(test_y_3, prediction_3)
print('Mean absolute error of random forest regression model #3: {}'.format(mae_3))
```
